## Visualize HCAS Policies
#### Inputs: 
* **nnetFile** - path to the .nnet file you want to visualize.
* **trainingDataFile** - path to training data file


#### Outputs: Interactive Plots

##### Interactive Settings:
* **psi**: Intruder relative heading angle
* **ownspeed**:   Ownship speed
* **intrspeed**:   Intruder speed
* **tau**: Time to loss of vertical separation
* **pra**: Previous RA
* **zoom**:    A zoom scaling factor used in plotting. 4 is most zoomed in, while 1 has no zoom
* **nbin**:    The number of bins being plotted. Higher nbin means better resolution, but that takes longer to compute
* **add_cocPenalty**: Add back the penalty removed to COC actions when pa is not 1 (COC). Default is false
* **add_online**: Add online costs to the table and network costs. Default is false
* **savePlot**:This option and the remaining options are used to format and generate .tex files of the plots. Set to true if you want to save the plot, but as a result the plots shown here will no longer continue to update. The switch can be toggled to true whenever you have a plot you want to save.
* **xshift**:  Shift the plot in the x-direction by a number of kft
* **yshift**:  Shift the plot in the y-direction by a number of kft
* **xscale**:  Stretch the plot in the x-direction by some scaling factor
* **yscale**:  Stretch the plot in the y-direction by some scaling factor

```julia
using PolicyViz, GridInterpolations, HDF5, Printf, PGFPlots

intTurnDelta = 1*pi/180
cocTurnDelta = 1*pi/180
grid  = RectangleGrid(RANGES,THETAS,PSIS,OWNSPEEDS,INTRSPEEDS)

function eval_table(Q,relState,grid)
    bel = get_belief(relState,grid,false)
    return Q[:,bel.rowval[1]]
end
function relativeState(own_state,intr_state)
    x0, y0, th0, v0 = own_state
    x1, y1, th1, v1 = intr_state
    r = sqrt((x0-x1).^2 + (y0-y1).^2)
    th = atan(y1-y0,x1-x0)-th0
    p  = th1 - th0
    
    th = th > pi  ? th-2*pi : th
    th = th < -pi ? th+2*pi : th
    p = p > pi  ? p-2*pi : p
    p = p < -pi ? p+2*pi : p
    
    return [r,th,p,v0,v1]
end

function dynamics(own_state,intr_state,adv,ignoreSelfSep,rng)
    turn1 = (2*rand(rng)-1.0)*intTurnDelta
    turn0 = (2*rand(rng)-1.0)*cocTurnDelta
    if (!ignoreSelfSep) .& (adv==2)
        turn0 += 1.5*pi/180
    elseif (!ignoreSelfSep) .& (adv==3)
        turn0 -= 1.5*pi/180
    elseif adv==4
        turn0 += 3.0*pi/180
    elseif adv==5
        turn0 -= 3.0*pi/180
    end
    x0, y0, th0, v0 = own_state
    x1, y1, th1, v1 = intr_state
    
    x0 += v0*cos(th0)
    y0 += v0*sin(th0)
    th0 += turn0
    
    x1 += v1*cos(th1)
    y1 += v1*sin(th1)
    th1 += turn1
    
    th0 = th0>pi ? th0-2*pi : th0
    th0 = th0<-pi ? th0 + 2*pi : th0
    th1 = th1>pi ? th1-2*pi : th1
    th1 = th1<-pi ? th1 + 2*pi : th1
    
    return [x0,y0,th0,v0], [x1,y1,th1,v1]
end
    
    
function simulateTraj(own_state,intr_state;randSeed=0,useTable=false,useCAS=true,nnetFile=nothing,grid=nothing,numT::Int=150,ignoreSelfSep=false)
    traj_own = zeros(2,numT+1)
    traj_int = zeros(2,numT+1)
    traj_own[:,1] = own_state[1:2]
    traj_int[:,1] = intr_state[1:2]
    
    rng = MersenneTwister(randSeed)
    pra = 0; tau=0
    adv=-1
    Q = []
    nnet = []
    if (!useTable) && (useCAS)
        nnet = read_network(nnetFile)
    end
    for t = 1:numT
        relState = relativeState(own_state,intr_state)
        if (useTable) && (useCAS)
            if pra!=adv-1
                Q = h5open(@sprintf("/raid/kjulian3/HCAS/TrainingData/HCAS_oneSpeed_TrainingData_v1_pra%d_tau%02d.h5",pra,0), "r") do file
                    read(file, "y")
                end
            end
            qvals = eval_table(Q,relState,grid)
            adv = findmax(qvals)[2]
        elseif useCAS
            
            qvals = evaluate_network(nnet,vcat(relState,[0,pra]))
            adv = findmax(qvals)[2]
        else
            adv=1
        end
        own_state,intr_state = dynamics(own_state,intr_state,adv,ignoreSelfSep,rng)
        traj_own[:,t+1] = own_state[1:2]
        traj_int[:,t+1] = intr_state[1:2]
        pra=adv-1
    end
    return traj_own, traj_int
end 

function getInitialStates(;randSeed=nothing)
    if randSeed==nothing
        randSeed = rand(1:100000)
    end
    rng = MersenneTwister(randSeed)
    own_state = [0,0,0,200.0]
    p_intr = (rand(rng)-0.5)*2*pi*5/6
    x_intr = 20000.0*(1+cos(p_intr))
    y_intr = 20000.0*sin(p_intr)
    p_intr+= pi
    p_intr = p_intr>pi ? p_intr-2*pi : p_intr
    p_intr = p_intr<-pi ? p_intr+2*pi : p_intr
    intr_state = [x_intr,y_intr,p_intr,200.0]
    
    return own_state, intr_state, randSeed
end

function getMinDistance(numSims)
    minDist_table = 1e10
    minDist_nnet = 1e10
    minDist_table_ind=-1
    minDist_nnet_ind=-1
    for i = 1:numSims
        own_state,intr_state,randSeed = getInitialStates(randSeed=i)
        traj_own_table, traj_int_table = simulateTraj(own_state,intr_state,randSeed=randSeed,grid=grid,useTable=true,numT=200,ignoreSelfSep=ignoreSelfSep);
        traj_own_nnet, traj_int_nnet = simulateTraj(own_state,intr_state,randSeed=randSeed,nnetFile=nnetFile,useTable=false,numT=200,ignoreSelfSep=ignoreSelfSep);
        dCPA_table,tCPA_table = findmin(sqrt.((traj_own_table[1,:]-traj_int_table[1,:]).^2 + (traj_own_table[2,:]-traj_int_table[2,:]).^2))
        dCPA_nnet,tCPA_nnet = findmin(sqrt.((traj_own_nnet[1,:]-traj_int_nnet[1,:]).^2 + (traj_own_nnet[2,:]-traj_int_nnet[2,:]).^2))
        
        if dCPA_table < minDist_table
            minDist_table = dCPA_table
            minDist_table_ind = i
        end
        if dCPA_nnet < minDist_nnet
            minDist_nnet = dCPA_nnet
            minDist_nnet_ind = i
        end
    end
    return [minDist_table_ind, minDist_table], [minDist_nnet_ind,minDist_nnet]
end

        
    
        

function plotRandomTraj(randSeed=nothing)
    own_state,intr_state,randSeed = getInitialStates(randSeed=randSeed)
    ignoreSelfSep = false
    nnetFile = "/raid/kjulian3/HCAS/networks/HCAS_oneSpeed_v1_45HU_200/HCAS_oneSpeed_v1_45HU_200.nnet"

    traj_own_table, traj_int_table = simulateTraj(own_state,intr_state,randSeed=randSeed,grid=grid,useTable=true,numT=200,ignoreSelfSep=ignoreSelfSep);
    traj_own_nnet, traj_int_nnet = simulateTraj(own_state,intr_state,randSeed=randSeed,nnetFile=nnetFile,useTable=false,numT=200,ignoreSelfSep=ignoreSelfSep);
    traj_own_nocas, traj_int_nocas = simulateTraj(own_state,intr_state,randSeed=randSeed,useCAS=false,numT=200);
    dCPA_table,tCPA_table = findmin(sqrt.((traj_own_table[1,:]-traj_int_table[1,:]).^2 + (traj_own_table[2,:]-traj_int_table[2,:]).^2))
    dCPA_nocas,tCPA_nocas = findmin(sqrt.((traj_own_nocas[1,:]-traj_int_nocas[1,:]).^2 + (traj_own_nocas[2,:]-traj_int_nocas[2,:]).^2))
    dCPA_nnet,tCPA_nnet = findmin(sqrt.((traj_own_nnet[1,:]-traj_int_nnet[1,:]).^2 + (traj_own_nnet[2,:]-traj_int_nnet[2,:]).^2))

    finalT = size(traj_own_nocas)[2]
    g = GroupPlot(3, 1, groupStyle = "horizontal sep=1cm")
    push!(g, Axis([
        Plots.Linear(traj_own_nocas[1,1:finalT],traj_own_nocas[2,1:finalT],mark="none",legendentry="Ownship"),
        Plots.Linear(traj_int_nocas[1,1:finalT],traj_int_nocas[2,1:finalT],mark="none",legendentry="Intruder"),
        Plots.Linear(traj_int_nocas[1,1:2],traj_int_nocas[2,1:2],mark="*",style="red, fill=red, mark size=3"),
        Plots.Linear(traj_own_nocas[1,1:2],traj_own_nocas[2,1:2],mark="*",style="blue, fill=blue, mark size=3")
        ],axisEqual=true,title="No CAS, dCPA: "*string(round(Int,dCPA_nocas))*"ft",legendPos="north west"))

    push!(g, Axis([
        Plots.Linear(traj_own_table[1,1:finalT],traj_own_table[2,1:finalT],mark="none",legendentry="Ownship"),
        Plots.Linear(traj_int_table[1,1:finalT],traj_int_table[2,1:finalT],mark="none",legendentry="Intruder"),
        Plots.Linear(traj_int_table[1,1:2],traj_int_table[2,1:2],mark="*",style="red, fill=red, mark size=3"),
        Plots.Linear(traj_own_table[1,1:2],traj_own_table[2,1:2],mark="*",style="blue, fill=blue, mark size=3")
        ],axisEqual=true,title="Table, dCPA: "*string(round(Int,dCPA_table))*"ft",legendPos="north west"))

    push!(g, Axis([
        Plots.Linear(traj_own_nnet[1,1:finalT],traj_own_nnet[2,1:finalT],mark="none",legendentry="Ownship"),
        Plots.Linear(traj_int_nnet[1,1:finalT],traj_int_nnet[2,1:finalT],mark="none",legendentry="Intruder"),
        Plots.Linear(traj_int_nnet[1,1:2],traj_int_nnet[2,1:2],mark="*",style="red, fill=red, mark size=3"),
        Plots.Linear(traj_own_nnet[1,1:2],traj_own_nnet[2,1:2],mark="*",style="blue, fill=blue, mark size=3")
        ],axisEqual=true,title="Network CAS, dCPA: "*string(round(Int,dCPA_nnet))*"ft",legendPos="north west"))
    return g
end
```
